## 3. Box Plots

```python
sns.boxplot(data = irisDataset, x = 'species', y = 'petal_length', hue = 'species')
```

```python
# Identifying outliers (computationally). I'm doing it for my own understanding of box plots, not as a part of this task. That's why there might be some slight mistakes (while showing outliers) in computations of mine (using pandas) and computations of seaborn internally. 

groups = irisDataset.groupby('species')

for species, group in groups:
    col = group['petal_length']

    Q1 = group['petal_length'].quantile(0.25)
    Q3 = group['petal_length'].quantile(0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    outliers = group[(group['petal_length'] < lower_bound) | group['petal_length'] > upper_bound]
    
    print(f"Outliers in {species}:\n {outliers}")
```
